```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
 df = pd. read_csv("Airbnb_data.csv")
```

# Handling missing values

```python
df['last review'] = pd.to_datetime(df['last review'], errors = 'coerce')   #coerce means if any date is missing, then it will show Not a thing 
```

```python
df.info()
```

```python
df.fillna( {' reviews per month' : 0, 'last review' : df['last review'].min()}, inplace = True)    #inplace = true means permanently change in data
```

```python
df.dropna(subset = ['NAME', 'host name'], inplace = True)
```

```python
print(df.isnull().sum())
```

```python
df = df.drop(columns=["license", "house_rules"], errors='ignore')
```

```python
df.head()
```

```python
 # Remove $ signs and convert to float
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
df['service fee'] = df['service fee'].replace('[\$,]', '', regex=True).astype(float)  #regex means to perform or execute
```

```python
df.head()
```

# remove duplicates

```python
df.drop_duplicates(inplace = True)   #inplace means to execute the change
```

```python
df.info()
```

# Visualization

# What is the distribution of listing prices?

```python
plt.figure(figsize = (10,6))
sns.histplot(df['price'], bins = 50, kde = True, color = 'red')      #kde shows a curve
plt.title(" Distribution of listing price")
plt.xlabel("Price $")
plt.ylabel("Frequency")
plt.show()
```

The histogram shows a fairly even distribution of 
listing prices across different price ranges , indicating no
particular concentration of listings in any specific price range,
The KDE line helps visualize this even spread more clearly, confirming that
the dataset contains listings with a wide variety of  prices.

```python
plt.figure(figsize = (8,5))
sns.countplot (x = 'room type', data = df, color = 'hotpink')
plt.title('room type distribution')
plt.xlabel('room type')
plt.ylabel('count')
plt.show()
```

# How are the listings distributed across different neighbouhoods?

```python
plt.figure(figsize = (12,8))
sns.countplot(y = 'neighbourhood group', data=df, color='lightgreen', order=df['neighbourhood group'].value_counts().index)
plt.title('Number of listings by Neighbourhood Group')
plt.xlabel('Count')
plt.ylabel('Neighbourhood Group')
plt.show()
```

# What is the relationahip and room type?

```python
plt.figure(figsize = (10,6 ))
sns.boxplot (x='room type', y='price', hue='room type',  data=df, palette='Set1')
plt.title('Price vs. Room Type')
plt.xlabel('Room type')
plt.ylabel('Price($)')
plt.legend(title='Room Type')
plt.show()
```

 # how has the number of reviews change over time?

```python
df['last review'] = pd.to_datetime(df['last review'])
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()
plt.figure(figsize = (12,6 ))
reviews_over_time.plot(kind='line', color='red')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()
```
